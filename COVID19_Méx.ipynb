```python
# Constante para el nombre del archivo
NOMBRE_ARCHIVO = 'casos_confirmados.csv'

# Función para analizar y mostrar una vista previa del archivo
def analizar_archivo(NOMBRE_ARCHIVO):
    """Analiza el archivo CSV y muestra encabezados y una muestra de datos."""
    print(f"\tAnalizando el archivo {NOMBRE_ARCHIVO}...")
    with open(NOMBRE_ARCHIVO, 'r', encoding='utf-8') as archivo:
        lineas = archivo.readlines()

        # Mostrar encabezados
        encabezados = lineas[0].strip().split(',')
        print("Encabezados del archivo:", encabezados)

        # Mostrar las primeras 3 filas de datos
        print("Primeras 3 filas de datos (muestra):")
        for linea in lineas[1:4]:
            print(linea.strip().split(','))
    print("\n\t--- Fin del análisis inicial ---\n")

# Función para procesar los datos y calcular estadísticas
def procesar_datos(NOMBRE_ARCHIVO):
    """Procesa el archivo CSV y devuelve edad promedio, conteo de géneros y moda de edades."""
    total_edad = 0
    num_pacientes = 0
    hombres = 0
    mujeres = 0
    frecuencias_edad = {}  # Diccionario para contar frecuencias de edades

    with open(NOMBRE_ARCHIVO, 'r', encoding='utf-8') as archivo:
        lineas = archivo.readlines()

        # Iterar sobre cada renglón, saltando el primero (encabezados)
        for linea in lineas[1:]:
            # Fragmentar la línea en columnas
            columnas = linea.strip().split(',')

            # Obtener y convertir la edad (columna 3)
            edad = int(columnas[3])

            # Obtener el género (columna 2)
            sexo = columnas[2]

            # Acumular datos
            total_edad += edad
            num_pacientes += 1

            # Contar hombres y mujeres
            if sexo == 'MASCULINO':
                hombres += 1
            elif sexo == 'FEMENINO':
                mujeres += 1

            # Contar frecuencia de la edad
            if edad in frecuencias_edad:
                frecuencias_edad[edad] += 1
            else:
                frecuencias_edad[edad] = 1

    # Calcular edad promedio
    edad_promedio = total_edad / num_pacientes if num_pacientes > 0 else 0

    # Calcular la moda
    if frecuencias_edad:
        max_frecuencia = max(frecuencias_edad.values())
        moda = [edad for edad, freq in frecuencias_edad.items() if freq == max_frecuencia]
    else:
        moda = []  # En caso de que no haya datos

    return edad_promedio, hombres, mujeres, moda

# Función para determinar la mayoría de género
def determinar_mayoria(hombres, mujeres):
    """Determina si hay más hombres o mujeres y retorna el resultado."""
    if hombres > mujeres:
        return "Hombres"
    elif mujeres > hombres:
        return "Mujeres"
    else:
        return "Igual número de hombres y mujeres"

# Función principal para ejecutar el programa
def main():
    """Función principal que coordina el análisis y procesamiento."""
    archivo = NOMBRE_ARCHIVO

    # Paso 1: Analizar el archivo
    analizar_archivo(archivo)

    # Pasos 3-5: Procesar datos
    edad_promedio, hombres, mujeres, moda = procesar_datos(archivo)

    # Determinar mayoría
    mayoria = determinar_mayoria(hombres, mujeres)

    # Paso 6: Imprimir resultados en formato claro
    print("\tRespuestas:")
    print(f"¿Cuál es la edad promedio de los pacientes? {edad_promedio:.2f} años")
    print(f"¿Hay más hombres o mujeres contagiados?")
    print(f"  - Hombres: {hombres}")
    print(f"  - Mujeres: {mujeres}")
    print(f"  - La mayoría son: {mayoria}")
    print("¿Cuál es la edad más frecuente en los casos confirmados?")
    if moda:
        print(f"- {', '.join(map(str, moda))} años")
    else:
        print("  - No hay datos para calcular la edad más frecuente")

# Ejecutar el programa
if __name__ == "__main__":
    main()
```
